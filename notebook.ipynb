---
Dataset ini berasal dari `Kaggle` dari link `https://www.kaggle.com/datasets/mubashirsidiki/student-academic-performance-500-students` <br>

| Nama Kolom              | Deskripsi                                      |
|-------------------------|------------------------------------------------|
| `student_id`            | ID unik student                                |
| `gender`                | Jenis kelamin (Wanita/Pria)                    |
| `age`                   | Umur / Usia                                    |
| `study_hours_per_week`  | Waktu belajar dalam satuan jam/minggu          |
| `attendent_rate`        | Rentang kehadiran dari 0 - 100                 |
| `parent_education`      | Tingkatan pendidikan orang tua                 |
| `internet_access`       | Apakah student memiliki akses internet (y/n)?  |
| `extracurricular`       | Apakah student memiliki ekstrakulikuler (y/n)? |
| `	previous_score`       | Nilai sebelumnya (UTS)                         |
| `	final_score`          | Nilai akhir (UAS)                              |
| `	passed`               | Apakah student lulus (y/n)? **Ini adalah label**|
