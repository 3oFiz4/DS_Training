## Import Library dan Data Loading


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
import kagglehub
import os

# Download latest version
path = kagglehub.dataset_download("mubashirsidiki/student-academic-performance-500-students")

print("Path to dataset files:", path)

# Construct the full path to the CSV file
file_path = os.path.join(path, "student_performance.csv")

df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Check apakah ada data yang missing pada dataset

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))

missing_percentage = (
    df.isnull().sum() / len(df) * 100
) # Mengambil persentase missing value

missing_percentage = missing_percentage[
    missing_percentage > 0
].sort_values()

ax = missing_percentage.plot.barh()

for p in ax.patches:
    percentage = p.get_width()
    count = p.get_y() + p.get_height()/2
    ax.annotate(
        f'{percentage:.1f}%',
        (
            p.get_x() + p.get_width()/2,
            count
        ),
        fontweight='bold'
    )

plt.xlabel(
    "Persentase Missing (%)",
    fontweight="bold",
    labelpad=10
    )
plt.title("Persentase Missing Value", fontsize=16, fontweight="bold", pad=20)
plt.show()
```

```python
# Check Distribusi Data Label (Passed)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))

ax = sns.countplot(
    x="passed",
    data=df,
    palette="Set1",
    hue="passed",
    legend=False
)
for p in ax.patches:
    count=p.get_height()

    ax.annotate(
        f'{count}',
        (
            p.get_x() + p.get_width()/2,
            count
        ),
        ha='center',
        fontweight="bold"
    )
ax.set_title(
    "Distribusi Label Passed",
    fontsize=16,
    fontweight="bold",
    pad=20
)
plt.show()
```

```python
# Visualisasi Distribusi Age

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))

ax = sns.countplot(
    x="age",
    data=df,
    hue="age",
    legend=False
)
for p in ax.patches:
    count=p.get_height()

    ax.annotate(
        f'{count}',
        (
            p.get_x() + p.get_width()/2,
            count
        ),
        ha='center',
        fontweight="bold"
    )
ax.set_title(
    "Distribusi Age",
    fontsize=16,
    fontweight="bold",
    pad=20
)
plt.show()
```
